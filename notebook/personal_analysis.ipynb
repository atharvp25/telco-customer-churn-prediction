```python
data=data.drop('customerID',axis=1)
```

```python
data['TotalCharges']=pd.to_numeric(data['TotalCharges'],errors='coerce')
data['TotalCharges'] = data['TotalCharges'].fillna(data['TotalCharges'].mean())
data['TotalCharges'].isnull().sum()
```

```python
# Create lists to store columns by category count
two_cat = []
three_cat = []
more_cat = []

# Loop through all object columns
for col in data.columns:
    if data[col].dtype == "O":
        unique_vals = data[col].unique()  # includes NaN as a unique value
        n_unique = len(unique_vals)
        
        if n_unique == 2:
            two_cat.append((col, unique_vals))
        elif n_unique == 3:
            three_cat.append((col, unique_vals))
        elif n_unique > 3:
            more_cat.append((col, unique_vals))

# Print results
if two_cat:
    print("Columns with Two Categories:")
    for col, vals in two_cat:
        print(f"{col} -> {vals}")
    print()

if three_cat:
    print("Columns with Three Categories:")
    for col, vals in three_cat:
        print(f"{col} -> {vals}")
    print()

if more_cat:
    print("Columns with More than Three Categories:")
    for col, vals in more_cat:
        print(f"{col} -> {vals}")
```

```python
# we check how many columns are same as it 
# extract only column names from three_cat list
three_cols = [col for col, vals in three_cat]

# compare each pair only once
for i in range(len(three_cols)):
    for j in range(i + 1, len(three_cols)):   # avoid duplicate & self comparisons

        col1 = three_cols[i]
        col2 = three_cols[j]

        if data[col1].equals(data[col2]):
            print(f"{col1} and {col2} are IDENTICAL columns.")
        else:
            print(f"{col1} and {col2} are NOT the same.")
```

```python
multi_categorical_cols = [
    "MultipleLines","InternetService","OnlineSecurity","OnlineBackup",
    "DeviceProtection","TechSupport","StreamingTV","StreamingMovies",
    "Contract","PaymentMethod"
]

data = pd.get_dummies(data, columns=multi_categorical_cols, drop_first=True, dtype=int)
```

```python
data.shape
```

```python
binary_cols = ["gender", "Partner", "Dependents", "PhoneService", "PaperlessBilling", "Churn"]

for col in binary_cols:
    data[col] = data[col].map({"Yes":1, "No":0, "Female":0, "Male":1})
```

```python
data.info()
```

```python
# Compute correlation
corr = data.corr()

# Correlation with Churn
corr_churn = corr["Churn"].sort_values(ascending=False)
print(corr_churn)
```

```python
# SO basics conclusion . 
'''
Tenure and contract type are the strongest predictors of churn.

Fiber optic + electronic check are strong signals for higher churn.

Features like gender, phone service, some streaming services are weak predictors.

Negative correlations for services (TechSupport, OnlineSecurity, OnlineBackup) suggest that offering these services reduces churn.'''
```

```python
from sklearn.feature_selection import SelectKBest, chi2

# X = features, y = target
X = data.drop("Churn", axis=1)
y = data["Churn"]

# Select top 15 features using chi-squared test
selector = SelectKBest(score_func=chi2, k=15)
X_new = selector.fit_transform(X, y)

# Get selected feature names
selected_features = X.columns[selector.get_support()].tolist()
print("Selected features:", selected_features)
```

```python
from sklearn.model_selection import train_test_split

X = data[selected_features]
y = data["Churn"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler

numeric_cols = ["tenure", "MonthlyCharges", "TotalCharges"]

scaler = StandardScaler()

# scale only numeric columns
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols]=scaler.transform(X_test[numeric_cols])

```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000, class_weight='balanced')

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
import numpy as np
from sklearn.metrics import classification_report

y_prob = model.predict_proba(X_test)[:,1]  # predicted probability of churn


for threshold in [0.3, 0.35, 0.4, 0.45, 0.5]:
    y_pred_new = (y_prob >= threshold).astype(int)
    print("Threshold =", threshold)
    print(classification_report(y_test, y_pred_new))
    print("--------")

```
