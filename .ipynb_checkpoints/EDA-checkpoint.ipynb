## Exploring the data categories wise

```python
colm = data.columns

for i in colm:
    if data[i].dtype == "O":
        print(i," -> ",data[i].unique())
```

```python
# Create lists to store columns by category count
two_cat = []
three_cat = []
more_cat = []

# Loop through all object columns
for col in data.columns:
    if data[col].dtype == "O":
        unique_vals = data[col].unique()  # includes NaN as a unique value
        n_unique = len(unique_vals)
        
        if n_unique == 2:
            two_cat.append((col, unique_vals))
        elif n_unique == 3:
            three_cat.append((col, unique_vals))
        elif n_unique > 3:
            more_cat.append((col, unique_vals))

# Print results
if two_cat:
    print("Columns with Two Categories:")
    for col, vals in two_cat:
        print(f"{col} -> {vals}")
    print()

if three_cat:
    print("Columns with Three Categories:")
    for col, vals in three_cat:
        print(f"{col} -> {vals}")
    print()

if more_cat:
    print("Columns with More than Three Categories:")
    for col, vals in more_cat:
        print(f"{col} -> {vals}")
```

### So we have 6 column which have TWO categories , then we have 9 column which have THREE categories , 
### then we have 1 column which have FOUR categories . Now we have to Apply one hot enconding on 10 columns and 
### Labeled Enconding on 6 columns 

```python
#so before encondind we check is there any column two or more column data are same or not .
three_cols = [col for col, vals in three_cat]

# compare each pair only once
for i in range(len(three_cols)):
    for j in range(i + 1, len(three_cols)):   # avoid duplicate & self comparisons

        col1 = three_cols[i]
        col2 = three_cols[j]

        if data[col1].equals(data[col2]):
            print(f"{col1} and {col2} are IDENTICAL columns.")
        else:
            print(f"{col1} and {col2} are NOT the same.")
```

### There is no single column are same so we have to encode each column 

```python
#Enconding Column which have more than two categories 
#Here we use One Hot Enconding 
multi_categorical_cols = [
    "MultipleLines","InternetService","OnlineSecurity","OnlineBackup",
    "DeviceProtection","TechSupport","StreamingTV","StreamingMovies",
    "Contract","PaymentMethod"
]

data = pd.get_dummies(data, columns=multi_categorical_cols, drop_first=True, dtype=int)
```

```python
#Ecoding Column which have two categories only 
#Here we use Label Enconding 
binary_cols = ["gender", "Partner", "Dependents", "PhoneService", "PaperlessBilling", "Churn"]

for col in binary_cols:
    data[col] = data[col].map({"Yes":1, "No":0, "Female":0, "Male":1})
```

## Here We Complete Data Preprocessing . In this we drop useless column , Clean data , Transform data , Encode Data
## Now We Are Ready for Further Analysis 

```python
# Here We are calculating corr so we see how many column are corr with each other 
# Compute correlation
corr = data.corr()

# Correlation with Churn (Churn is Our Target)
corr_churn = corr["Churn"].sort_values(ascending=False)
print(corr_churn)
```

```python
#Basics Conclusion 

#Tenure and contract type are the strongest predictors of churn.
#Fiber optic + electronic check are strong signals for higher churn.
#Features like gender, phone service, some streaming services are weak predictors.
#Negative correlations for services (TechSupport, OnlineSecurity, OnlineBackup) suggest that offering these services reduces churn.
```

```python
#So We have 31 features and now we have to reduce it and pick up top 15 feature which are help us for model 
```

```python
#Feature Engineering
from sklearn.feature_selection import SelectKBest, chi2

# X = features, y = target
X = data.drop("Churn", axis=1)
y = data["Churn"]

# Select top 15 features using chi-squared test
selector = SelectKBest(score_func=chi2, k=15)
X_new = selector.fit_transform(X, y)

# Get selected feature names
selected_features = X.columns[selector.get_support()].tolist()
print("Selected features:", selected_features)
```

```python
#Spliting the data in trainging and testing 
from sklearn.model_selection import train_test_split

X = data[selected_features]
y = data["Churn"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
#Checking dimension
#X_train 
#X_test
#y_test
#y_train
```

```python
# Standardlized the numeric Features 
from sklearn.preprocessing import StandardScaler

numeric_cols = ["tenure", "MonthlyCharges", "TotalCharges"]

scaler = StandardScaler()

# scale only numeric columns
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols]=scaler.transform(X_test[numeric_cols])
```

```python
#Checking is Our Column is Imbalance or Not 
data['Churn'].value_counts()
```

```python
#Our Target Feature is Imblance so while traning the model we have to Balance it 
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000, class_weight='balanced')

model.fit(X_train, y_train)
```

```python
#Prediction for X_test
y_pred = model.predict(X_test)
```

```python
#Accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#We have 74% of Accuracy
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
# For Better Result we set Threshold 
# Usually its 0.5 but our case (Imbalance Dataset) we have to set this via some values .
```

```python
import numpy as np
from sklearn.metrics import classification_report

y_prob = model.predict_proba(X_test)[:,1]  # predicted probability of churn

#Here we take some common threshold and try on our model and check which fit better 
for threshold in [0.3, 0.35, 0.4, 0.45, 0.5]:
    y_pred_new = (y_prob >= threshold).astype(int)
    print("Threshold =", threshold)
    print(classification_report(y_test, y_pred_new))
    print("--------")

```

```python
#Amoung All these threshold we select 0.4 becz 


#Precision ≈ 0.47 → reasonable number of correct churn predictions
#Recall ≈ 0.91 → most churners are detected
#F1-score ≈ 0.62 → overall balanced performance
#Lower threshold (0.3) → recall slightly higher, but precision too low (many false positives)
#Higher threshold (0.45–0.5) → precision higher, but recall drops (miss many churners)
```

```python
# Sample2 Input And Prediction

feature_columns = [
    'tenure', 'MonthlyCharges', 'TotalCharges',
    'InternetService_Fiber optic', 'InternetService_No',
    'OnlineSecurity_No internet service', 'OnlineSecurity_Yes',
    'OnlineBackup_No internet service', 'DeviceProtection_No internet service',
    'TechSupport_No internet service', 'StreamingTV_No internet service',
    'StreamingMovies_No internet service', 'Contract_One year',
    'Contract_Two year', 'PaymentMethod_Electronic check'
]

user_input = [
    72,        # tenure: customer for 6 years
    50.0,      # MonthlyCharges: moderate
    3500.0,    # TotalCharges: high total payment
    0, 1,      # InternetService_Fiber optic, InternetService_No
    0, 0,      # OnlineSecurity columns (No internet service / Yes)
    0,         # OnlineBackup_No internet service
    0,         # DeviceProtection_No internet service
    0,         # TechSupporret_No internet service
    0,         # StreamingTV_No internet service
    0,         # StreamingMovies_No internet service
    0, 1,      # Contract_One year, Contract_Two year
    0          # PaymentMethod_Electronic check
]

# Convert to DataFrame (1 row)
X_input = pd.DataFrame([user_input], columns=feature_columns)

# Scale numeric columns if needed
numeric_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
X_input[numeric_cols] = scaler.transform(X_input[numeric_cols])  # if you used scaler

# Predict probability
y_prob = model.predict_proba(X_input)[:,1]

# Apply threshold 0.4
threshold = 0.4
y_pred = (y_prob >= threshold).astype(int)

# Show result
if y_pred[0] == 1:
    print(f"Prediction: Churn (Probability={y_prob[0]:.2f})")
else:
    print(f"Prediction: No Churn (Probability={y_prob[0]:.2f})")
```

```python
# Define the same feature columns as used in training
feature_columns = [
    'tenure', 'MonthlyCharges', 'TotalCharges',
    'InternetService_Fiber optic', 'InternetService_No',
    'OnlineSecurity_No internet service', 'OnlineSecurity_Yes',
    'OnlineBackup_No internet service', 'DeviceProtection_No internet service',
    'TechSupport_No internet service', 'StreamingTV_No internet service',
    'StreamingMovies_No internet service', 'Contract_One year',
    'Contract_Two year', 'PaymentMethod_Electronic check'
]

# Function to convert human-readable input into one-hot encoded input
def encode_input(user_dict):
    # Start with all zeros
    x = np.zeros(len(feature_columns))
    
    # Numeric features
    x[feature_columns.index('tenure')] = user_dict['tenure']
    x[feature_columns.index('MonthlyCharges')] = user_dict['MonthlyCharges']
    x[feature_columns.index('TotalCharges')] = user_dict['TotalCharges']
    
    # InternetService
    if user_dict['InternetService'] == 'Fiber optic':
        x[feature_columns.index('InternetService_Fiber optic')] = 1
    elif user_dict['InternetService'] == 'No':
        x[feature_columns.index('InternetService_No')] = 1
    # DSL is the baseline → drop_first
    
    # OnlineSecurity
    if user_dict['OnlineSecurity'] == 'Yes':
        x[feature_columns.index('OnlineSecurity_Yes')] = 1
    elif user_dict['OnlineSecurity'] == 'No internet service':
        x[feature_columns.index('OnlineSecurity_No internet service')] = 1
    # 'No' is baseline
    
    # OnlineBackup
    if user_dict['OnlineBackup'] == 'No internet service':
        x[feature_columns.index('OnlineBackup_No internet service')] = 1
    # 'Yes' is baseline (or you can adjust depending on drop_first)
    
    # DeviceProtection
    if user_dict['DeviceProtection'] == 'No internet service':
        x[feature_columns.index('DeviceProtection_No internet service')] = 1
    
    # TechSupport
    if user_dict['TechSupport'] == 'No internet service':
        x[feature_columns.index('TechSupport_No internet service')] = 1
    
    # StreamingTV
    if user_dict['StreamingTV'] == 'No internet service':
        x[feature_columns.index('StreamingTV_No internet service')] = 1
    
    # StreamingMovies
    if user_dict['StreamingMovies'] == 'No internet service':
        x[feature_columns.index('StreamingMovies_No internet service')] = 1
    
    # Contract
    if user_dict['Contract'] == 'One year':
        x[feature_columns.index('Contract_One year')] = 1
    elif user_dict['Contract'] == 'Two year':
        x[feature_columns.index('Contract_Two year')] = 1
    # Month-to-month is baseline
    
    # PaymentMethod
    if user_dict['PaymentMethod'] == 'Electronic check':
        x[feature_columns.index('PaymentMethod_Electronic check')] = 1
    # Other payment methods are baseline
    
    return pd.DataFrame([x], columns=feature_columns)

# Example human-readable input
user_dict = {
    'tenure': 12,
    'MonthlyCharges': 75.5,
    'TotalCharges': 910.5,
    'InternetService': 'DSL',
    'OnlineSecurity': 'Yes',
    'OnlineBackup': 'Yes',
    'DeviceProtection': 'No internet service',
    'TechSupport': 'No internet service',
    'StreamingTV': 'No internet service',
    'StreamingMovies': 'No internet service',
    'Contract': 'One year',
    'PaymentMethod': 'Electronic check'
}

# Encode input
X_input = encode_input(user_dict)

# Predict probability
y_prob = model.predict_proba(X_input)[:,1]

# Apply threshold 0.4
threshold = 0.4
y_pred = (y_prob >= threshold).astype(int)

if y_pred[0] == 1:
    print(f"Prediction: Churn (Probability={y_prob[0]:.2f})")
else:
    print(f"Prediction: No Churn (Probability={y_prob[0]:.2f})")
```
